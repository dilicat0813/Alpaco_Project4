```python
import dash
import dash_core_components as dcc
import dash_bootstrap_components as dbc
import dash_html_components as html
from dash.dependencies import Input, Output
from jupyter_dash import JupyterDash



app = JupyterDash(__name__)


app.layout = dbc.Container(
    [
        html.Img(id='image', src='https://img.hankyung.com/photo/202111/PEP20200720164101055_P4.jpg'),
        dbc.Button('바셀 메시', id='btn-1', n_clicks=0, className='mr-1'),
        dbc.Button('아르헨 메시', id='btn-2', n_clicks=0, className='mr-2'),
        dbc.Button('파리 메시', id='btn-3', n_clicks=0, className='mr-3'),
    ],
    fluid=False,
)

# @app.callback(
#     Output('image', 'src'),
#     [Input('btn-1', 'n_clicks'),
#      Input('btn-2', 'n_clicks'),
#      Input('btn-3', 'n_clicks')],
#     [State('btn-1', 'n_clicks_timestamp'),
#      State('btn-2', 'n_clicks_timestamp'),
#      State('btn-3', 'n_clicks_timestamp')]
# )
# def update_image(btn1, btn2, btn3, ts1, ts2, ts3):
#     # 버튼 클릭 시간을 비교하여 마지막으로 클릭된 버튼을 판별
#     last_clicked = max(ts1, ts2, ts3)
#     if last_clicked == ts1:
#         return 'https://blog.kakaocdn.net/dn/cjm1ao/btrbor1V4M4/M79GDry75P87v6EiC22pD1/img.png'
#     elif last_clicked == ts2:
#         return 'https://newsimg-hams.hankookilbo.com/2022/10/18/5e8cbea0-b91e-4774-b5e0-3060e1b0e4cb.jpg'
#     elif last_clicked == ts3:
#         return 'https://image.newsis.com/2023/02/19/NISI20230219_0019789249_web.jpg'
#     else:
#         return 'https://img.hankyung.com/photo/202111/PEP20200720164101055_P4.jpg'

@app.callback(
    Output('image', 'src'),
    [Input('btn-1', 'n_clicks'),
     Input('btn-2', 'n_clicks'),
     Input('btn-3', 'n_clicks')],
    [State('btn-1', 'n_clicks_timestamp'),
     State('btn-2', 'n_clicks_timestamp'),
     State('btn-3', 'n_clicks_timestamp')],
)
def update_image(btn1, btn2, btn3, ts1, ts2, ts3):
    ts1 = ts1 or 0  # 클릭 시간이 NoneType인 경우 기본값 0으로 설정
    ts2 = ts2 or 0
    ts3 = ts3 or 0

    last_clicked = max(ts1, ts2, ts3)
    if last_clicked == ts1:
        return 'https://blog.kakaocdn.net/dn/cjm1ao/btrbor1V4M4/M79GDry75P87v6EiC22pD1/img.png'
    elif last_clicked == ts2:
        return 'https://newsimg-hams.hankookilbo.com/2022/10/18/5e8cbea0-b91e-4774-b5e0-3060e1b0e4cb.jpg'
    elif last_clicked == ts3:
        return 'https://image.newsis.com/2023/02/19/NISI20230219_0019789249_web.jpg'
    else:
        return 'https://img.hankyung.com/photo/202111/PEP20200720164101055_P4.jpg'


if __name__ == '__main__':
    app.run_server(debug=True)
```
